# Stemming

```python
# Começando a realizar o pré-processamento
# Stemming em Portugues
# É uma técnica utilizada para reduzir a palavra na sua forma "radical". 
# Criando objetio stemmer.
stemmer = nltk.stem.RSLPStemmer()
```

```python
stemmer.stem('Trabalhou')
```

```python
stemmer.stem('correu')
```

```python
# Função para aplicar Stemming
def Stemming(instancia):
    stemmer = nltk.stem.RSLPStemmer() # Cria um objeto do tipo Stemming.
    palavras=[] # Cria uma variável do tipo lista.
    for w in instancia.split(): # Loop para "partir" a instância.
        palavras.append(stemmer.stem(w))# Adiciona a palavra na lista aplicando o stem.
    return (" ".join(palavras)) # Realiza um join no resultando. Concatenando com um "espaço".
```

```python
Stemming('Ele trabalhava frequentemente no projeto de criacao de novos produtos')
```

# Lemmatization

```python
# Lemmatizarion para pré-processamento de dados.
# Desflexiona uma palavra para determinar o seu "Lemma"
from nltk.stem import WordNetLemmatizer
```

```python
# Criando um objeto tipo Lemmatizarion
lemmatizer = WordNetLemmatizer()
```

```python
# Aplicando o lemmatize na palavra "cats"
print (lemmatizer.lemmatize("cats"))
```

```python
# Busca um lemma para 'is' do tipo verbo
# O parâmetro pos é que define o tipo de busca que será realizado.
lemmatizer.lemmatize('is', pos='v')
```

```python
# Busca um adjetivo para 'better'
print (lemmatizer.lemmatize("better", pos="a"))
```

#  Stopwords

```python
# Stopwords são palavras ou termos muito usados mas que não tem nenhum significado. Ex: Artigos e, o, a....
# Técnica de pré-processamento de dados.
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
# Exibe quais são as Stopwords do conjunto que está sendo utilizado.
stopwords
```

```python
# Criação de uma função para remover as Stopwords
def RemoviStopWords(instancia):
    instancia = instancia.lower() # Transforma em minúsculo e atribui o resultado na variável instancia.
    stopwords = set(nltk.corpus.stopwords.words('portuguese')) # Set para retornar apenas uma ocorrência da palavra.
    palavras = [i for i in instancia.split() if not i in stopwords] # Loop para guardar as palavras que não são stopwords
    return (" ".join(palavras))# Junção com um espaço
```

```python
sentenca1 = 'Ola mundo, estou trabalhando com textos com o artigo do minerando dados'
sentenca2 = 'Seu trabalho de NLP será super interessante com as aulas de Mineração de Textos'
```

```python
RemoviStopWords(sentenca2)
```

# Part of Speech

```python
# Part of Speech: Na tardução literal seria: Partes da Fala.
# Consistem em identificar a **função gramatical** das palavras em um texto.
# A função gramatical muda conforme é empregada a palavra no texto.

text = nltk.word_tokenize('O blog Minerando Dados é muito bom')
nltk.pos_tag(text)
```
